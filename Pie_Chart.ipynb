## Pie charts are used to show part-to-whole relationships. They're visually appealing but best used with fewer categories, as too many slices can get cluttered and hard to interpret.

```python
import matplotlib.pyplot as plt

plt.style.available
```

```python
plt.style.use("ggplot")
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.pie(runs, labels = labels)
plt.show()
```

```python
colors = ["#1F3A93","#22A7F0","#26C281","#F5AB35"]
plt.pie(runs, labels = labels, colors = colors)
plt.show()
```

```python
plt.pie(runs,
       labels = labels,
       colors = colors,
       wedgeprops = {'edgecolor' : 'black','linestyle' : '--','linewidth' : 2, 'fill' : True}
)
plt.show()
```

```python
plt.pie(runs,
       labels = labels,
       colors = colors,
       wedgeprops = {'edgecolor' : 'black','linestyle' : '--','linewidth' : 2, 'fill' : False}
)
plt.show()
```

```python
plt.pie(runs,
       labels = labels,
       colors = colors,
       wedgeprops = {'edgecolor' : 'red','linestyle' : '--','linewidth' : 2}
)
plt.show()
```

# Highlight a Slice Using explode
## Use explode to pull out one or more slices.

```python
explode = [0.1,0,0,0] #Only Highlight Sachin's Slic
plt.pie(runs,labels = labels, colors = colors,explode = explode)
plt.show()
```

```python
explode = [0,0,0,0.2]
plt.pie(runs,labels = labels, colors = colors,explode = explode)
plt.show()
```

```python
plt.pie(runs,labels = labels, colors = colors,explode = explode, shadow = True)
plt.show()
```

# Start Angle
## Use startangle to rotate the pie for better alignment.

```python
plt.pie(runs,labels = labels, colors = colors, shadow = True,explode = [0,0,0,0.2],startangle = 140)
plt.show()
```

```python
plt.pie(runs,labels = labels, colors = colors, shadow = True,explode = [0,0,0,0.2],startangle = 10)
plt.show()
```

# Show Percentages with autopct

```python
plt.pie(runs,colors = colors,labels = labels,shadow = True,startangle = 10, autopct = "%1.1f%%")
plt.show()
```

# Crowded Pie Chart  Why to Avoid

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```

# Solution

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]
a = 0
others_sum_usage = sum([i for i in usage if i < 10])
usage_filtered = [i for i in usage if i >=10]
languages_filtered = [languages[i] for i,v in enumerate(usage) if v >=10]
usage_filtered.append(others_sum_usage)
languages_filtered.append("Others")

print(usage_filtered)
print(languages_filtered)

    

plt.pie(usage_filtered, labels=languages_filtered, autopct='%1.1f%%', startangle=90,  )
plt.title("Programming Language Usage (Crowded Example)")
plt.text(1,1,"Others ='C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP'", color = "darkblue")
plt.show()
```

## Why avoid it?

### Hard to compare slice sizes visually
### Cluttered and confusing
### No clear insight
### Better alternatives: bar charts or horizontal bar charts
