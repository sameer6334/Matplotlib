```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years,runs)
plt.xlabel("years")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

# Mulyiple Lines in one plot

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli ,label = "Virat")
plt.plot(years,sehwag, label = "Sehwag")
plt.xlabel("years")
plt.ylabel("run scored")
plt.legend()
plt.title("Kohli's and Virat's run during many Years")
plt.show()
```

# using format strings

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years,kohli, "ro--", label = "Virat")
plt.plot(years, sehwag, "g^:", label = "sehwag")
plt.legend()
plt.xlabel("years")
plt.ylabel("runs scored")
plt.title("Kohli's and Virat's run during many Years")
plt.show()
```

```python
plt.plot(years,kohli_runs, color = "red", linestyle = "--", label = "kohli")
plt.plot(years,rohit_runs, color = "blue", linestyle = "-.", label = "rohit")
plt.plot(years,sehwag_runs, color = "yellow", linestyle = "-",label = "sehwag")
plt.legend()
plt.style.use("seaborn-v0_8-paper")
plt.grid("True")
plt.tight_layout()
```
