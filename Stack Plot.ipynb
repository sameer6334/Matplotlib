```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]  # hours in a day

colors = ["red","yellow","green","pink"]
plt.figure(figsize = (6,6))
plt.pie(time_spent,labels = activities,colors = colors, autopct = "%1.1f%%", startangle = 90)
plt.title("school going students")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# What is a Stack Plot?

## Now, let’s imagine you surveyed multiple days, and you want to track how the time spent on each activity changes over a week.

## A Stack Plot is a type of area chart that helps visualize multiple quantities over time, stacked on top of each other. It's especially useful to see how individual parts contribute to a whole over time.

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['studying','playing','watching_tv','sleeping']
colors = ["red","yellow","green","pink"]
plt.figure(figsize = (6,6))
plt.stackplot(days,studying,playing,watching_tv,sleeping, labels = labels, colors = colors,alpha = 0.4)
plt.legend(loc = 'upper left')
plt.title("students time distribution")
plt.xlabel("days")
plt.ylabel("hours")
plt.grid(True)

plt.show()
```

```python
# Days of the week
days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]  # 1 = Monday, 7 = Sunday

# Hours spent on activities each day (sum = 24)
python_learning = [3, 3, 2, 3, 3, 2, 2]
solving_questions = [2, 2, 3, 2, 2, 2, 2]
video_editing = [2, 1, 2, 2, 2, 2, 1]
content_management = [2, 2, 1, 2, 2, 2, 2]
ml_study = [2, 1, 2, 1, 1, 1, 2]
resting_sleeping = [13, 15, 14, 14, 14, 15, 15]  # total hours to complete 24
colors = ['#1f77b4',  # Blue
          '#ff7f0e',  # Orange
          '#2ca02c',  # Green
          '#d62728',  # Red
          '#9467bd',  # Purple
          '#8c564b']  # Brown
labels = ['python_learning','solving_questions','video_editing','content_management','ml_study','resting_sleeping']
plt.stackplot(days,python_learning,solving_questions,video_editing,content_management,ml_study,resting_sleeping, colors = colors, labels = labels, alpha = 0.5)
plt.legend(loc = "best")
plt.title("number Hours I spend In a day per week")
plt.xlabel("days of the week")
plt.ylabel("Hours")
plt.show()
```
